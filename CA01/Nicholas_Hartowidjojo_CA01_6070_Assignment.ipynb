```python
data = house_price_train_df[['OverallQual', 'SalePrice']].dropna()

groups = []
labels = []

for q in sorted(data['OverallQual'].unique()):
    groups.append(data[data['OverallQual'] == q]['SalePrice'])
    labels.append(str(q))

plt.figure(figsize=(8,4))
plt.boxplot(groups, labels=labels)
plt.title("SalePrice by Overall Quality")
plt.xlabel("OverallQual")
plt.ylabel("SalePrice")
plt.tight_layout()
plt.show()
```

```python
top_neigh = house_price_train_df['Neighborhood'].value_counts().head(8).index
data = house_price_train_df[house_price_train_df['Neighborhood'].isin(top_neigh)][['Neighborhood', 'SalePrice']]

groups = []
labels = []

for n in top_neigh:
    groups.append(data[data['Neighborhood'] == n]['SalePrice'])
    labels.append(n)

plt.figure(figsize=(10,4))
plt.boxplot(groups, labels=labels)
plt.title("SalePrice by Neighborhood (Top 8)")
plt.xlabel("Neighborhood")
plt.ylabel("SalePrice")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
table = pd.crosstab(house_price_train_df['Neighborhood'], house_price_train_df['HouseStyle'])

top_neigh = house_price_train_df['Neighborhood'].value_counts().head(8).index
table = table.loc[top_neigh]

table.plot(kind='bar', figsize=(10,4))
plt.title("HouseStyle Distribution by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
num_cols = df_clean.select_dtypes(include='number').columns

# fill missing numeric with median
for col in num_cols:
    if df_clean[col].isna().sum() > 0:
        median_value = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_value)
```

```python
#find the missing values in the clean copy dataset
df_clean.isna().sum().sum()
```

```python
numeric_cols = house_price_train_df.select_dtypes(include='number').columns
numeric_cols = [c for c in numeric_cols if c != 'SalePrice']
```

```python
outlier_summary = []

for col in numeric_cols:
    data = house_price_train_df[col].dropna()

    if data.nunique() < 10:
        continue

    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = ((house_price_train_df[col] < lower_bound) |
                     (house_price_train_df[col] > upper_bound)).sum()

    outlier_summary.append([col, outlier_count])
```

```python
#detecting outliers
outlier_df = pd.DataFrame(
    outlier_summary,
    columns=['column', 'outlier_count']
).sort_values('outlier_count', ascending=False)

outlier_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ambil fitur numerik saja
numeric_df = df_encoded.select_dtypes(include='number')

# hitung korelasi
corr = numeric_df.corr()

plt.figure(figsize=(12, 10))
plt.imshow(corr, aspect='auto')
plt.colorbar(label='Correlation')
plt.title("Correlation Heatmap (Numeric Features)")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.index)), corr.index)
plt.tight_layout()
plt.show()
```

```python
threshold = 0.8

high_corr_pairs = []

for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > threshold:
            high_corr_pairs.append((
                corr.columns[i],
                corr.columns[j],
                corr.iloc[i, j]
            ))

high_corr_pairs[:10]
```
